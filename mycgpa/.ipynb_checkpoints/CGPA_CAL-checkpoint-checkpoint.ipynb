```python
import numpy as pd
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import  LinearRegression
```

```python
df=pd.read_csv("cgpa.csv")
```

```python
df.head()
```

```python
df.first_year.median()
```

```python
df.first_year=df.first_year.fillna(df.first_year.median())
df.second_year=df.second_year.fillna(df.second_year.median())
df.third_year=df.third_year.fillna(df.third_year.median())
df.fourth_year=df.fourth_year.fillna(df.fourth_year.median())
df
```

```python
X=df.drop('result',axis=1)
X
```

```python
Y=df[['result']]
Y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=2)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,y_pred))
```
